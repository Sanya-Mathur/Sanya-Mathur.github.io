```python
# Import necessary libraries
import pandas as pd
from scipy.stats import f_oneway

# Mount Google Drive if you want to upload your file there (optional)
# from google.colab import drive
# drive.mount('/content/drive')

# Load the Excel file
# If uploading manually through Colab's interface, use this approach
from google.colab import files
uploaded = files.upload()

# Load the uploaded Excel file
  # Get the uploaded file's name
data = pd.read_excel("/content/final entries of bs.xlsx")

# Define mappings to convert categorical responses into numeric values
response_mapping_experience = {
    'Strongly Agree': 5, 'Agree': 4, 'Neutral': 3, 'Disagree': 2, 'Strongly Disagree': 1
}
response_mapping_disruptive = {
    'Very Disruptive': 5, 'Somewhat Disruptive': 4, 'Neutral': 3, 'Not Disruptive': 2, 'Not at All Disruptive': 1
}
response_mapping_willingness = {
    'Yes': 1, 'No': 0
}
response_mapping_sustainability = {
    'Strongly Agree': 5, 'Agree': 4, 'Neutral': 3, 'Disagree': 2, 'Strongly Disagree': 1
}

# Map the numeric values to the relevant columns
data['ad_experience_rating'] = data['10. On a scale of 1 to 5, how effective do you think advertising models are for balancing revenue generation with user experience? (1 being least effective, 5 being most effective)']
data['subscription_experience_rating'] = data['5. Do you think a subscription model leads to a better overall user experience compared to free or ad-supported platforms?'].map(response_mapping_experience)
data['ad_disruptive_rating'] = data['2. How disruptive do you find advertisements on platforms that you use?'].map(response_mapping_disruptive)
data['ad_willingness_rating'] = data['3. Are you willing to watch ads in exchange for free services or content?'].map(response_mapping_willingness)
data['sustainability_rating'] = data['9. Do you believe that ad-supported models are more sustainable for long-term growth than freemium or subscription models?'].map(response_mapping_sustainability)

# Drop rows with missing values in any of the relevant columns for ANOVA
data_cleaned = data.dropna(subset=[
    'ad_experience_rating',
    'subscription_experience_rating',
    'ad_disruptive_rating',
    'ad_willingness_rating',
    'sustainability_rating'
])

# Perform the ANOVA tests

# 1. ANOVA for User Experience Ratings: comparing ad vs. subscription experience
anova_experience = f_oneway(
    data_cleaned['ad_experience_rating'],
    data_cleaned['subscription_experience_rating']
)
print("ANOVA for User Experience Ratings (ad vs. subscription):", anova_experience)

# 2. ANOVA for Sustainability Perceptions: comparing sustainability perceptions of ad-supported vs. subscription models
anova_sustainability = f_oneway(
    data_cleaned['sustainability_rating'],
    data_cleaned['subscription_experience_rating']
)
print("ANOVA for Sustainability Perceptions (ad vs. subscription):", anova_sustainability)

# 3. ANOVA for Ad Tolerance Ratings: comparing disruptiveness of ads and willingness to view ads for free content
anova_ad_tolerance = f_oneway(
    data_cleaned['ad_disruptive_rating'],
    data_cleaned['ad_willingness_rating']
)
print("ANOVA for Ad Tolerance Ratings (disruptiveness vs. willingness to view ads):", anova_ad_tolerance)
```

```python
# Import necessary libraries
import pandas as pd
import scipy.stats as stats

# Load the data (you can upload a file in Colab using the file upload feature)
from google.colab import files
uploaded = files.upload()

# Read the uploaded file (adjust the filename if needed)
file_path = list(uploaded.keys())[0]
data = pd.read_excel(file_path)  # Use read_csv if the file is in CSV format

# Prepare data for Chi-Square test
# Replace 'Gender' and 'Preference' with the actual column names in your dataset
chi_square_data = data[['Gender:', '1. How often do you choose freemium platforms over subscription or advertising-based ones?']].dropna()

# Create a contingency table
contingency_table = pd.crosstab(
    chi_square_data['Gender:'],
    chi_square_data['1. How often do you choose freemium platforms over subscription or advertising-based ones?']
)

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)
print("\nExpected Frequencies:")
print(expected)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Create the dataset as a dictionary
data = {
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Choice': ['Always', 'Always', 'Never', 'Never', 'Often', 'Often', 'Rarely', 'Rarely', 'Sometimes', 'Sometimes'],
    'Frequency': [5, 23, 2, 17, 16, 28, 3, 23, 12, 28]
}

# Convert it to a DataFrame
df = pd.DataFrame(data)

# Descriptive statistics: Count, Mean, Standard Deviation, etc.
descriptive_stats = df.groupby(['Gender', 'Choice'])['Frequency'].describe()

# Display the descriptive statistics
print("Descriptive Statistics:")
print(descriptive_stats)

# Set the style for the plot
sns.set(style="whitegrid")

# Create a bar plot of the frequency of each choice by gender
plt.figure(figsize=(10,6))
sns.barplot(x='Choice', y='Frequency', hue='Gender', data=df, palette='Set2')

# Add labels and title
plt.title('Frequency of Choosing Freemium Platforms by Gender', fontsize=14)
plt.xlabel('Freemium Platform Choice', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()

# Create a pivot table to reshape the data for a stacked bar chart
pivot_df = df.pivot_table(values='Frequency', index='Choice', columns='Gender', aggfunc='sum')

# Plotting the stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(10,6), color=['lightblue', 'lightgreen'])

# Add labels and title
plt.title('Stacked Bar Chart of Freemium Platform Choices by Gender', fontsize=14)
plt.xlabel('Freemium Platform Choice', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()

# Pie chart to show the gender distribution
gender_counts = df['Gender'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(7,7))
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'], legend=False)

# Add title
plt.title('Gender Distribution in the Dataset', fontsize=14)

# Show the plot
plt.show()

# Create a contingency table for the Chi-Square test
contingency_table = pd.crosstab(df['Choice'], df['Gender'])

# Perform the Chi-Square test
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)

# Display the Chi-Square Test results
print("\nChi-Square Test Results:")
print(f'Chi-Square Statistic: {chi2_stat}')
print(f'p-value: {p_val}')
print(f'Degrees of Freedom: {dof}')
print(f'Expected Frequencies: \n{ex}')
```
