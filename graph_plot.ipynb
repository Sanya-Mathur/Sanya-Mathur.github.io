```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'East'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'B'],
    'Sales': [100, 200, 150, 250, 120, 300, 180],
    'Profit': [20, 50, 40, 60, 30, 80, 70]
}
df = pd.DataFrame(data)

# Line Plot
df.groupby('Region')[['Sales', 'Profit']].sum().plot(kind='line', title='Line Plot')
plt.show()

# Stacked Bar Plot
df.groupby('Region')[['Sales', 'Profit']].sum().plot(kind='bar', stacked=True, title='Stacked Bar Plot')
plt.show()

# Horizontal Bar Plot
df.groupby('Region')[['Sales', 'Profit']].sum().plot(kind='barh', title='Horizontal Bar Plot')
plt.show()

# Box Plot
df[['Sales', 'Profit']].plot(kind='box', title='Box Plot')
plt.show()

# Horizontal Box Plot
df[['Sales', 'Profit']].plot(kind='box', vert=False, title='Horizontal Box Plot')
plt.show()

# Area Plot
df.groupby('Region')[['Sales', 'Profit']].sum().plot(kind='area', stacked=True, title='Area Plot')
plt.show()

# Scatter Plot
df.plot(kind='scatter', x='Sales', y='Profit', title='Scatter Plot')
plt.show()

# Histogram - Stacked Cumulative
df[['Sales', 'Profit']].plot(kind='hist', bins=10, stacked=True, cumulative=True, title='Stacked Cumulative Histogram', alpha=0.6)
plt.show()

# Pie Chart
df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', title='Pie Chart')
plt.ylabel('')  # Hide default ylabel
plt.show()

# Histogram of Each Column
df.hist(bins=10, figsize=(10, 5), alpha=0.7)
plt.suptitle('Histograms of Each Column')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'East'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'B'],
    'Sales': [100, 200, 150, 250, 120, 300, 180],
    'Profit': [20, 50, 40, 60, 30, 80, 70],
    'Discount': [5, 10, 7, 12, 6, 15, 8]
}
df = pd.DataFrame(data)

# Univariate Analysis: Analyze a single column
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Sales'], kde=True, bins=10, color='blue')
plt.title('Univariate Analysis: Sales Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Profit'], color='green')
plt.title('Univariate Analysis: Profit Box Plot')
plt.show()

# Bivariate Analysis: Analyze relationships between two columns
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Region', style='Product', palette='Set2')
plt.title('Bivariate Analysis: Sales vs Profit')

plt.subplot(1, 2, 2)
sns.barplot(data=df, x='Region', y='Discount', hue='Product', ci=None)
plt.title('Bivariate Analysis: Discount by Region and Product')
plt.show()

# Multivariate Analysis: Analyze relationships among three or more variables
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Multivariate Analysis: Correlation Heatmap')
plt.show()

# Pairplot for multivariate analysis
sns.pairplot(df, hue='Region', diag_kind='kde', palette='husl')
plt.suptitle('Multivariate Analysis: Pairplot', y=1.02)
plt.show()
```
